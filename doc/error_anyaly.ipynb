```python
def error_label_count(arrdt):
    if arrdt['prediction_labels'] == 'make up':
        comp = re.compile(r'\b(makeup|make up)\b')
    elif arrdt['prediction_labels'] == 'skin care':
        comp = re.compile(r'\b(skin care)\b')
    elif arrdt['prediction_labels'] == 'bodycare':
        comp = re.compile(r'\bbodycare\b')
    elif arrdt['prediction_labels'] == 'fashion':
        comp = re.compile(r'\bfashion\b')
    else:
        return 0
    
    r1 = comp.findall(arrdt['text'])
    
    if len(r1) > 0:
        counts = len(r1)
    else:
        counts = 0
    return counts

error_pd['kw_counts'] = error_pd.apply(error_label_count, axis=1)
```
